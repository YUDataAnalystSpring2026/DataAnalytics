**Week 6 — Functions with Records**

This notebook analyzes a small sales dataset using Python functions.
Each function takes the full dataset as its first argument, so it can be
called with different datasets without changing the function itself.

```python
sales_data = [
    ('Laptop',     'East',  1200.00),
    ('Phone',      'West',   850.00),
    ('Tablet',     'East',   400.00),
    ('Monitor',    'South',  650.00),
    ('Keyboard',   'West',    75.00),
    ('Laptop',     'South', 1150.00),
    ('Phone',      'East',   900.00),
    ('Tablet',     'West',   425.00),
    ('Monitor',    'East',   680.00),
    ('Keyboard',   'South',   80.00),
    ('Laptop',     'West',  1300.00)
]
```

## Section 1 — Basic Aggregations

```python
### Lab 1. Total sales by region
```

```python
# ALTERNATE 1 - uses a generator expression instead of a for loop
def region_total(data, region):
    """Return the sum of sale amounts for the given region."""
    return sum(amt for _, r, amt in data if r == region)

# ALTERNATE 2 - uses a for loop instead of a generator expression 
# def region_total(data, region):
    # """Return the sum of sale amounts for the given region."""
    # total = 0
    # for _, r, amt in data:
    #     if r == region:
    #         total += amt
    # return total
```

```python
print(f'East: {region_total(sales_data, 'East')}')
print(f'West: {region_total(sales_data, 'West')}')
print(f'South: {region_total(sales_data, 'South')}')
```

### Lab 2. Count of sales by region

```python
# ALTERNATE 1 - uses a generator expression instead of a for loop
def region_count(data, region):
    '''Return number of records for given region by summing +1 per record)'''
    return sum(1 for _, r, amt in data if r == region)

# ALTERNATE 2 - uses a for loop instead of a generator expression 
# def region_count(data, region):
#     '''Return the count of sales for the given region.'''
#     count = 0
#     for _, r, amt in data:
#         if r == region:
#             count += 1
#     return count
```

```python
print(f'East: {region_count(sales_data, 'East')}')
print(f'West: {region_count(sales_data, 'West')}')
print(f'South: {region_count(sales_data, 'South')}')
```

### Lab 3. Average sale amount by region

```python
def region_average(data, region):
    '''Returns average sales for a given region, or 'Not found' if region doesn't exist'''
    count = region_count(data, region)
    if count == 0:
        return 'Not found'
    return f'{(region_total(data, region) / count):,.2f}'
```

```python
print(region_average(sales_data, 'East'))
print(region_average(sales_data, 'West'))
print(region_average(sales_data, 'South'))
print(region_average(sales_data, 'Midwest'))
```

## Section 2 — Filtering and Finding Extremes

### Lab 4. Filter by sale amount

```python
# The long way
# def sales_above(data, region, threshold):
#     '''Returns records where amount of sale is greater than the given threshold -- also includes filtering on region'''
#     sales_list = []
#     for _, r, amt in data:
#         if r == region:
#             if amt > threshold:
#                 sales_list.append((_, r, amt))
#     return sales_list

# Much shorter!
def sales_above(data, threshold):
    '''Returns records where amount of sale is greater than the given threshold'''
    return [s for s in data if s[2] > threshold]
```

```python
print(sales_above(sales_data, 400))
print(sales_above(sales_data, 50))
print(sales_above(sales_data, 1000))
print(sales_above(sales_data, 40000))
```

### Lab 5. Top and bottom sales

```python
def top_sale(data):
    """Return the record with the highest sale amount."""
    return max(data, key=lambda s: s[2])

def bottom_sale(data):
    """Return the record with the lowest sale amount."""
    return min(data, key=lambda s: s[2])
```

```python
print("--- Lab 5: top_sale and bottom_sale ---")
best = top_sale(sales_data)
worst = bottom_sale(sales_data)
print(f"Top sale: ${best[2]:,.2f} for a {best[0]} (Region: {best[1]})")
print(f"Bottom sale: ${worst[2]:,.2f} for a {worst[0]} (Region: {worst[1]})")
```

## Section 3 - Transforming Records

### Lab 6. Apply a discount

```python
def apply_discount(data, discount_pct):
    """Return a new list of records with sale amounts reduced by discount_pct.

    Args:
        data: list of (product, region, amount) tuples
        discount_pct: discount as a decimal, e.g. 0.15 for 15%
    """
    return list(map(lambda s: (s[0], s[1], s[2] * (1 - discount_pct)), data))
```

```python
print("--- Lab 6: apply_discount(15%) ---")
discounted = apply_discount(sales_data, 0.15)
print("First 3 records before discount:")
for record in sales_data[:3]:
    print(f"  {record[0]:<10} ${record[2]:,.2f}")
print("First 3 records after 15% discount:")
for record in discounted[:3]:
    print(f"  {record[0]:<10} ${record[2]:,.2f}")
```

### Lab 7. Unique regions

```python
def unique_regions(data):
    """Return a sorted list of unique region names in the dataset."""
    return sorted(set(r for _, r, _ in data))
```

```python
print("--- Lab 7: unique_regions ---")
print(unique_regions(sales_data))
```

## Section 4 — Putting It Together

### Lab 8. Region summary

```python
def region_summary(data, region):
    """Print a formatted summary report for the given region."""
    regional_data = [s for s in data if s[1] == region]

    if not regional_data:
        print(f"=== {region} Region ===")
        print("  No records found for this region.")
        return

    total   = region_total(data, region)
    count   = region_count(data, region)
    average = region_average(data, region)
    best    = top_sale(regional_data)  # top sale within this region only

    print(f"=== {region} Region ===")
    print(f"  Records:      {count}")
    print(f"  Total Sales:  ${total:,.2f}")
    print(f"  Average Sale: ${average}")
    print(f"  Top Sale:     {best[0]} — ${best[2]:,.2f}")
```

```python
print("--- Lab 8: region_summary ---")
region_summary(sales_data, 'East')
print()
region_summary(sales_data, 'West')
print()
region_summary(sales_data, 'South')
print()
region_summary(sales_data, 'North')  # not in data — should handle gracefully
```

### Lab 9. Full Summary (Challenge)

```python
print("--- Lab 9: Full summary (all regions) ---")
for region in unique_regions(sales_data):
    region_summary(sales_data, region)
    print()
```
