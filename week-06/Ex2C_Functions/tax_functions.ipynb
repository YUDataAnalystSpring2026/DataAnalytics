```python
def get_soc_sec_tax(gross_pay):
    return gross_pay * .062
```

```python
def get_medicare_tax(gross_pay):
    return gross_pay * .0145
```

```python
def get_federal_tax(gross_pay, withh):
    if withh == 0:
        return gross_pay * .23
    elif withh == 1:
        return gross_pay * .21
    elif withh == 2:
        return gross_pay * .195
    elif withh == 3:
        return gross_pay * .185
    elif withh >= 4:
        return gross_pay * .18
    else:
        return "Entry not valid"
```

```python
# Person1

print(f'''
    For gross pay of {750}, withholding code {0}:
    Social Security tax = {get_soc_sec_tax(750)}
    Medicare tax = {get_medicare_tax(750)}
    Federal tax = {get_federal_tax(750, 0)}
''')
```
